```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df=pd.read_csv('/content/alcohol_sales (1).csv',
               parse_dates=['Date'],
               index_col='Date')
```

```python
df.head()
```

```python
plt.plot(df['Sales'])
```

Ce graphique presente une allure croissante de 1992  2020.
On en dduit que les ventes d'Alcool dans les supermarchs en Amerique augmentent au fil des annes

```python
#Find the series decomposition
decomposed=seasonal_decompose(df['Sales'])
```

```python
decomposed.plot()
```



Le graphique de dcomposition visualise la rpartition des donnes de sries chronologiques (df['Sales']) en quatre composants:

Observed: il s'agit des donnes de sries chronologiques d'origine des ventes d'alcool vue precdemment .

Trend: ce composant reprsente le modle global  long terme des donnes. Dans ce cas, il montre une tendance  la hausse, indiquant une augmentation des ventes d'alcool au fil du temps.

Saisonal: ce composant capture les modles ou fluctuations rptitifs qui se produisent au cours d'une priode fixe, comme une anne dans ce cas. Nous pouvez voir des pics et des creux rguliers dans ce composant, mettant en vidence la saisonnalit des ventes d'alcool.

Rsid: ce composant reprsente la variation restante des donnes aprs prise en compte de la tendance et de la saisonnalit. Il devrait idalement ressembler  un bruit alatoire, indiquant que le modle a captur les principaux modles.

Interprtation:

En examinant ces composants ensemble, nous pouvons obtenir des informations sur la dynamique des ventes d'alcool. La tendance  la hausse suggre une augmentation gnrale des ventes au fil des ans. La composante saisonnire rvle des modles rcurrents, peut-tre dus aux vacances, aux saisons ou  d'autres facteurs influenant le comportement des consommateurs. Le graphique rsiduel nous aide  valuer dans quelle mesure le modle s'adapte aux donnes. S'il prsente un bruit alatoire, cela indique une bonne adquation. Les modles restants dans les rsidus peuvent indiquer des informations non captures ou des limites du modle.


```python
# Analyse residual
sns.distplot(decomposed.resid)
```

En observant le graphique , nous pouvons voir que la distribution des residus ressemble  une distribution normale. Cela indique que le modle a bien captur les principaux modle de la serie temporelle
