```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df=pd.read_csv('/content/alcohol_sales (1).csv',
               parse_dates=['Date'],
               index_col='Date')
```

```python
df.head()
```

```python
plt.plot(df['Sales'])
```

Ce graphique presente une allure croissante de 1992 à 2020.
On en déduit que les ventes d'Alcool dans les supermarchés en Amerique augmentent au fil des années

```python
#Find the series decomposition
decomposed=seasonal_decompose(df['Sales'])
```

```python
decomposed.plot()
```



Le graphique de décomposition visualise la répartition des données de séries chronologiques (df['Sales']) en quatre composants :

Observed : il s'agit des données de séries chronologiques d'origine des ventes d'alcool vue precédemment .

Trend : ce composant représente le modèle global à long terme des données. Dans ce cas, il montre une tendance à la hausse, indiquant une augmentation des ventes d'alcool au fil du temps.

Saisonal : ce composant capture les modèles ou fluctuations répétitifs qui se produisent au cours d'une période fixe, comme une année dans ce cas. Nous pouvez voir des pics et des creux réguliers dans ce composant, mettant en évidence la saisonnalité des ventes d'alcool.

Résid : ce composant représente la variation restante des données après prise en compte de la tendance et de la saisonnalité. Il devrait idéalement ressembler à un bruit aléatoire, indiquant que le modèle a capturé les principaux modèles.

Interprétation :

En examinant ces composants ensemble, nous pouvons obtenir des informations sur la dynamique des ventes d'alcool. La tendance à la hausse suggère une augmentation générale des ventes au fil des ans. La composante saisonnière révèle des modèles récurrents, peut-être dus aux vacances, aux saisons ou à d'autres facteurs influençant le comportement des consommateurs. Le graphique résiduel nous aide à évaluer dans quelle mesure le modèle s'adapte aux données. S'il présente un bruit aléatoire, cela indique une bonne adéquation. Les modèles restants dans les résidus peuvent indiquer des informations non capturées ou des limites du modèle.


```python
# Analyse residual
sns.distplot(decomposed.resid)
```

En observant le graphique , nous pouvons voir que la distribution des residus ressemble à une distribution normale. Cela indique que le modèle a bien capturé les principaux modèle de la serie temporelle
