```python

# la somme des décès pour chaque cause de mortalité
cause_of_death_sums = df.drop(columns=["Country", "Year"]).sum()

# Trier les causes de décès par ordre décroissant
cause_of_death_sums = cause_of_death_sums.sort_values(ascending=False)
# Afficher les résultats
print(f"Nombre de cause de mortalités : {len(cause_of_death_sums)}")
print("")
print(cause_of_death_sums)

# Créer le bar plot
plt.figure(figsize=(15, 9))  # Ajuster la taille de la figure si nécessaire
plt.bar(cause_of_death_sums.index, cause_of_death_sums.values)
plt.xticks(rotation=90, ha='right')  # Faire pivoter les étiquettes de l'axe x pour une meilleure lisibilité
plt.title("Principales causes de mortalité")
plt.xlabel("Cause de décès")
plt.ylabel("Nombre de décès")
plt.tight_layout()  # Ajuster l'espacement pour éviter le chevauchement des éléments
plt.show()
```
